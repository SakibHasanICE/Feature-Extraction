```python
import pandas as pd
import numpy as np 
```

```python
df_n=df
df_s=df
df_l=df
df_r=df
df_m=df
df2=df
```

```python
df_n[['Administration']] = m.fit_transform(df[['Administration']])
df_n.head()
```

```python
for (columnName, columnData) in df_n.iteritems():    
    if(df_n[columnName].dtype == 'float64' and columnName != 'Profit'):
        df_n[[columnName]] = m.fit_transform(df_n[[columnName]])
df_n.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stds = StandardScaler()
```
